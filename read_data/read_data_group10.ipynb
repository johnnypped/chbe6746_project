This file reads the corresponding pickle files for inputs (atomic positions and atomic fingerprints) and the output (energies in eV). 

```python
import pickle
import numpy as np

# loading numpy arrays from pickle files
positions_pkl = "positions.pkl"
fingerprints_pkl = "fingerprints.pkl"
energies_pkl = "energies.pkl"

with open(positions_pkl, "rb") as f:
    positions = pickle.load(f)

with open(fingerprints_pkl, "rb") as f:
    fingerprints = pickle.load(f)

with open(energies_pkl, "rb") as f:
    energies = pickle.load(f)

# print the shape and the items for the first data point

print(positions.shape) # 400 for total data points, 3 as for 3 atoms, 3 for x,y,z respectively
print(fingerprints.shape) # 400 for total data points, 3 for 3 atoms, 80 dimensions for fingerprints
print(energies.shape) # one energy output per data point


print(positions[0])
print(fingerprints[0])
print(energies[0])
```
